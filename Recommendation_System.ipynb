```python
movies = movies.merge(credits,on='title')
movies.head()
```

```python
movies = movies[['movie_id','title','cast','crew','genres','keywords','overview']]
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
import ast
def convert(text):
  m=[]
  for i in ast.literal_eval(text):
    m.append(i['name'])
  return m
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[0]['genres']
```

```python
movies.iloc[0]['keywords']
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.iloc[0]['keywords']
```

```python
def convert_cast(text):
  m=[]
  counter=0
  for i in ast.literal_eval(text):
    if counter<=3:
      m.append(i['name'])
    counter+=1
  return m
```

```python
movies.iloc[2]['cast']
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.iloc[2]['cast']
```

```python
movies.head()
```

```python
def director(text):
  m=[]
  for i in ast.literal_eval(text):
    if i['job'] == 'Director':
      m.append(i['name'])
      break
  return m
```

```python
movies.iloc[3]['crew']
```

```python
movies['crew'] = movies['crew'].apply(director)
```

```python
movies.iloc[3]['crew']
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply( lambda x: x.split())
movies['overview']
```

```python
def removespace(text):
  m=[]
  for i in text:
    m.append(i.replace(" ",''))
  return m
```

```python
movies['cast'] = movies['cast'].apply(removespace)
movies['crew'] = movies['crew'].apply(removespace)
movies['keywords'] = movies['keywords'].apply(removespace)
movies['genres'] = movies['genres'].apply(removespace)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.iloc[9]['tags']
```

```python
movies.head()
```

```python
newdf = movies[['movie_id','title','tags']]
newdf.head()
```

```python
newdf.iloc[23]['tags']
```

```python
newdf['tags'] = newdf['tags'].apply( lambda x:" ".join(x))
newdf['tags'] = newdf['tags'].apply( lambda x:x.lower())
newdf.head()
```

NLP WORKS

```python
import nltk
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stems(text):
  m = []
  for i in text.split():
    m.append(ps.stem(i))
  return ' '.join(m)
```

```python
newdf['tags'] = newdf['tags'].apply(stems)
newdf.iloc[9]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vector = cv.fit_transform(newdf['tags']).toarray()
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
newdf[newdf['title'] == 'Batman'].index[0]
```

```python
newdf[newdf['title'] == 'The Lego Movie'].index[0]
```

```python
def recommended(movie):
  index = newdf[newdf['title'] == movie].index[0]
  distance = sorted(list(enumerate(similarity[index])),reverse = True,key = lambda x:x[1])
  for i in distance[1:9]:
    print(newdf.iloc[i[0]]['title'])
```

```python
import pickle
import streamlit as st
import requests

def fetch_poster(movie_id):
    url = "https://api.themoviedb.org/3/movie/{}?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US".format(movie_id)
    data = requests.get(url)
    data = data.json()
    poster_path = data['poster_path']
    full_path = "https://image.tmdb.org/t/p/w500/" + poster_path
    return full_path

def recommended(movie):
    index = movies[movies['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    recommended_movie_names = []
    recommended_movie_posters = []
    for i in distances[1:6]:
        movie_id = movies.iloc[i[0]].movie_id
        recommended_movie_posters.append(fetch_poster(movie_id))
        recommended_movie_names.append(movies.iloc[i[0]].title)

    return recommended_movie_names,recommended_movie_posters



st.header('Movie Recommender System Using Machine Learning')

movies = pickle.load(open("D:/BearandBee/r/saved/movie.sav",'rb'))
similarity = pickle.load(open("D:/BearandBee/r/saved/simili.sav",'rb'))

movie_list = movies['title'].values
selected_movie = st.selectbox(
    "Type or select a movie from the dropdown",
    movie_list
)

if st.button('Show Recommendation'):
    recommended_movie_names,recommended_movie_posters = recommended(selected_movie)
    col1, col2, col3, col4, col5 = st.columns(5)
    with col1:
        st.text(recommended_movie_names[0])
        st.image(recommended_movie_posters[0])
    with col2:
        st.text(recommended_movie_names[1])
        st.image(recommended_movie_posters[1])

    with col3:
        st.text(recommended_movie_names[2])
        st.image(recommended_movie_posters[2])
    with col4:
        st.text(recommended_movie_names[3])
        st.image(recommended_movie_posters[3])
    with col5:
        st.text(recommended_movie_names[4])
        st.image(recommended_movie_posters[4])

print(type(similarity))
```
